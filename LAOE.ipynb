# Introduction

This code solves four viscoelastic constitutive models (FENE-P, simplified PTT, Giesekus, and Oldroyd-B) in conformation tensor form (where $\tens{A}$ is the conformation tensor) under homogeneous oscillatory extensional flow. The flow field for a homogeneous oscillatory extensional flow is given by 

\begin{equation}
 \vect{\nabla}\vect{u} = \cos(t)\begin{bmatrix} 
 \displaystyle{\frac{-1}{2}}(1+\vartheta) & 0 & 0 \\
 0 & \displaystyle{\frac{-1}{2}}(1-\vartheta) & 0 \\
 0 & 0 & 1
 \end{bmatrix},
 \tag{1}
\end{equation}

where $\vartheta = 0$ for planar extensional flow, and $\vartheta = 1$ for Uniaxial ($\cos(t)>1$) and Biaxial ($\cos(t)<1$) extensional flow.

The Oldroyd-B model under oscillatory extensional flow is given by

\begin{equation}
    \frac{\dif A_{xx}}{\dif t} = -\frac{Wi}{De}(1+\vartheta)A_{xx}\cos(t) - \frac{1}{De}\left(A_{xx}-1\right)
    \tag{2a}
\end{equation}
\begin{equation}
    \frac{\dif A_{yy}}{\dif t} = -\frac{Wi}{De}(1-\vartheta)A_{yy}\cos(t) - \frac{1}{De}\left(A_{yy}-1\right)
    \tag{2b}
\end{equation}
\begin{equation}
    \frac{\dif A_{zz}}{\dif t} = 2\frac{Wi}{De}A_{yy}\cos(t) - \frac{1}{De}\left(A_{zz}-1\right)
    \tag{2c}
\end{equation}

The FENE-P model under oscillatory extensional flow is given by
\begin{equation}
    \frac{\dif A_{xx}}{\dif t} = -\frac{Wi}{De}(1+\vartheta)A_{xx}\cos(t) - \frac{1}{De}\left( \frac{L^2}{L^2-A_{xx}-A_{yy}-A_{zz}}A_{xx}-\frac{L^2}{L^2-3} \right),
    \tag{3a}
\end{equation}
\begin{equation}
    \frac{\dif A_{yy}}{\dif t} = -\frac{Wi}{De}(1-\vartheta)A_{yy}\cos(t) - \frac{1}{De}\left( \frac{L^2}{L^2-A_{xx}-A_{yy}-A_{zz}}A_{yy}-\frac{L^2}{L^2-3} \right),
    \tag{3b}
\end{equation}
\begin{equation}
    \frac{\dif A_{zz}}{\dif t} = 2\frac{Wi}{De}A_{zz}\cos(t) - \frac{1}{De}\left( \frac{L^2}{L^2-A_{xx}-A_{yy}-A_{zz}}A_{zz}-\frac{L^2}{L^2-3} \right).
    \tag{3c}
\end{equation}

The sPTT model under oscillatory extensional flow is given by

\begin{equation}
    \frac{\dif A_{xx}}{\dif t} = -\frac{Wi}{De}(1+\vartheta)A_{xx}\cos(t) - \frac{1}{De}[1+\epsilon(A_{xx}+A_{yy}+A_{zz}-3)]\left(A_{xx}-1\right),
    \tag{4a}
\end{equation}
\begin{equation}
    \frac{\dif A_{yy}}{\dif t} = -\frac{Wi}{De}(1-\vartheta)A_{yy}\cos(t) - \frac{1}{De}[1+\epsilon(A_{xx}+A_{yy}+A_{zz}-3)]\left(A_{yy}-1\right),
    \tag{4b}
\end{equation}
\begin{equation}
    \frac{\dif A_{zz}}{\dif t} = 2\frac{Wi}{De}A_{yy}\cos(t) - \frac{1}{De}[1+\epsilon(A_{xx}+A_{yy}+A_{zz}-3)]\left(A_{zz}-1\right). 
    \tag{4c}
\end{equation}

The Giesekus model under oscillatory extensional flow is given by

\begin{equation}
    \frac{\dif A_{xx}}{\dif t} = -\frac{Wi}{De}(1+\vartheta)A_{xx}\cos(t) - \frac{1}{De}[1+\alpha(A_{xx}-1)]\left(A_{xx}-1\right)
    \tag{5a}
\end{equation}
\begin{equation}
    \frac{\dif A_{yy}}{\dif t} = -\frac{Wi}{De}(1-\vartheta)A_{yy}\cos(t) - \frac{1}{De}[1+\alpha(A_{yy}-1)]\left(A_{yy}-1\right)
    \tag{5b}
\end{equation}
\begin{equation}
    \frac{\dif A_{zz}}{\dif t} = 2\frac{Wi}{De}A_{yy}\cos(t) - \frac{1}{De}[1+\alpha(A_{zz}-1)]\left(A_{zz}-1\right)
    \tag{5c}
\end{equation}

$De$ is the Deborah number defined by $De=\lambda\omega$ where $\omega$ is rotational frequency, and $Wi$ is the Weissenberg number defined by $Wi \equiv \lambda \epsilon_0\omega$ where $\epsilon_0$ is the strain amplitude. Note that time is non-dimensionalised by $\omega$ and $\vect{\nabla}\vect{u}$ is non-dimensionalised by $\epsilon_0\omega$ (or the strain-rate amplitude $\dot{\epsilon}_0$). Since we set $L^2 = 1/\epsilon = \alpha$, we define a parameter $\psi$ such that $\psi \equiv L^2 = 1/\epsilon = \alpha$ for brevity.

For the sPTT, Giesekus, and Oldroyd-B models, the polymer stress $\tens{\tau}_p$ is recovered by 

\begin{equation}
    \tens{\tau}_p = \frac{1}{Wi}(\tens{A}-\tens{I}),
    \tag{6}
\end{equation}

whereas for the FENE-P model, $\tens{\tau}_p$ is recovered by 

\begin{equation}
    \tens{\tau}_p = \frac{1}{Wi}\left(\frac{L^2}{L^2 - \tr \ \tens{A}}\tens{A}- \frac{L^2}{L^2-3}\tens{I}\right).
    \tag{7}
\end{equation}

Note here that we assume that there is no Newtonian solvent contribution to the extra stress tensor $\tens{\tau}$ and so $\tens{\tau} = \tens{\tau}_p$. 

The code below solves Equations 2-5 using the solve_ivp function with initial consitions $\tens{A}=\tens{I}$ and plots the viscous Lissajous curve for $N_1$, where $N_1 \equiv \tau_{xx}-\tau_{yy}$

```python
# Importing packages etc

import numpy as np
import matplotlib.colors as colors
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from matplotlib.ticker import FormatStrFormatter

from collections import OrderedDict

linestyles = OrderedDict(
    [('solid',               (0, ())),
     ('loosely dotted',      (0, (1, 10))),
     ('dotted',              (0, (1, 5))),
     ('densely dotted',      (0, (1, 1))),

     ('loosely dashed',      (0, (5, 10))),
     ('dashed',              (0, (5, 5))),
     ('densely dashed',      (0, (5, 1))),

     ('loosely dashdotted',  (0, (3, 10, 1, 10))),
     ('dashdotted',          (0, (3, 5, 1, 5))),
     ('densely dashdotted',  (0, (3, 1, 1, 1))),

     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))])
```

```python
# Defining functions which return the LHS of the equations above for the respective constitutive model

def fenep(t, y, Wi, De, theta, L2):
    trA = y[0]+y[1]+y[2]
    f = L2/(L2-trA)
    a = L2/(L2-3)
    return ([-(Wi/De)*(1+theta)*y[0]*np.cos(t)-(1/De)*(f*y[0]-a),
             -(Wi/De)*(1-theta)*y[1]*np.cos(t)-(1/De)*(f*y[1]-a),
             2*(Wi/De)*y[2]*np.cos(t)-(1/De)*(f*y[2]-a)])

def sptt(t, y, Wi, De, theta, epsilon):
    trA = y[0]+y[1]+y[2]
    f = 1+epsilon*(trA-3)
    return ([-(Wi/De)*(1+theta)*y[0]*np.cos(t)-(1/De)*f*(y[0]-1),
             -(Wi/De)*(1-theta)*y[1]*np.cos(t)-(1/De)*f*(y[1]-1),
             2*(Wi/De)*y[2]*np.cos(t)-(1/De)*f*(y[2]-1)])

def giesekus(t, y, Wi, De, theta, alpha):
    return ([-(Wi/De)*(1+theta)*y[0]*np.cos(t)-(1/De)*(1+alpha*(y[0]-1))*(y[0]-1),
             -(Wi/De)*(1-theta)*y[1]*np.cos(t)-(1/De)*(1+alpha*(y[1]-1))*(y[1]-1),
             2*(Wi/De)*y[2]*np.cos(t)-(1/De)*(1+alpha*(y[2]-1))*(y[2]-1)])

def oldroyd(t, y, Wi, De, theta):
    return ([-(Wi/De)*(1+theta)*y[0]*np.cos(t)-(1/De)*(y[0]-1),
             -(Wi/De)*(1-theta)*y[1]*np.cos(t)-(1/De)*(y[1]-1),
             2*(Wi/De)*y[2]*np.cos(t)-(1/De)*(y[2]-1)])
```

```python
# Making arrays of De and Wi - just to make it easy to cycle through them.
De_array = [0.001, 0.01, 0.1, 1, 10]
Wi_array = [0.125, 0.5, 2, 8]

# Choosing De and Wi values from arrays above.
i=3
j=3

De = De_array[i]
Wi = Wi_array[j]

# Defnining the flow type (0 for planar, 1 for uniaxial/biaxial).
theta = 0

# Choosing De and Wi values from arrays above.
psi = 100

# Specifying the number of periods and number of evaluation points per period.
Periods = 10
PointsPer = 300
t_end = Periods*2*np.pi


# Specifying solve_ivp options.
odemethod = 'LSODA'
rtol_val = 1E-8
atol_val = 1E-8




# General set up.
L2 = psi
epsilon = 1/L2
a = L2/(L2-3)
alpha = epsilon

tspan = np.linspace(2*np.pi/PointsPer,t_end,PointsPer*Periods)

################ FENE-P ##################

sol_fene = solve_ivp(fenep,(0, t_end),[1, 1, 1],method=odemethod,rtol=rtol_val,atol=atol_val,t_eval=tspan,args=(Wi/a,De/a,theta,L2))

trA_fene = sol_fene.y[0]+sol_fene.y[1]+sol_fene.y[2]
F_fene = np.divide(L2,L2-trA_fene)
a_fene = L2/(L2-3)

tau_11_fene = (a/Wi)*(np.multiply(F_fene,sol_fene.y[0])-a_fene)
tau_22_fene = (a/Wi)*(np.multiply(F_fene,sol_fene.y[1])-a_fene)
tau_33_fene = (a/Wi)*(np.multiply(F_fene,sol_fene.y[2])-a_fene)

N1_fene = tau_33_fene-tau_11_fene

################ FENE-P ##################


################ sPTT ##################

sol_sptt = solve_ivp(sptt,(0, t_end),[1, 1, 1],method=odemethod,rtol=rtol_val,atol=atol_val,t_eval=tspan,args=(Wi,De,theta,epsilon))

tau_11_sptt = (1/Wi)*(sol_sptt.y[0]-1)
tau_22_sptt = (1/Wi)*(sol_sptt.y[1]-1)
tau_33_sptt = (1/Wi)*(sol_sptt.y[2]-1)

N1_sptt = tau_33_sptt-tau_11_sptt

################ sPTT ##################

################ Giesekus ##################

sol_gs = solve_ivp(giesekus,(0, t_end),[1, 1, 1],method=odemethod,rtol=rtol_val,atol=atol_val,t_eval=tspan,args=(Wi,De,theta,alpha))

tau_11_gs = (1/Wi)*(sol_gs.y[0]-1)
tau_22_gs = (1/Wi)*(sol_gs.y[1]-1)
tau_33_gs = (1/Wi)*(sol_gs.y[2]-1)

N1_gs = tau_33_gs-tau_11_gs

################ Giesekus ##################

################ Oldroyd-B ##################

sol_ob = solve_ivp(oldroyd,(0, t_end),[1, 1, 1],method=odemethod,rtol=rtol_val,atol=atol_val,t_eval=tspan,args=(Wi,De,theta))

tau_11_ob = (1/Wi)*(sol_ob.y[0]-1)
tau_22_ob = (1/Wi)*(sol_ob.y[1]-1)
tau_33_ob = (1/Wi)*(sol_ob.y[2]-1)

N1_ob = tau_33_ob-tau_11_ob

################ Oldroyd-B ##################



################# Figures ####################

plt.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 6})
plt.rcParams["axes.axisbelow"] = False
cm = 1/2.54  # centimeters in inches
linewidth_value = 1.5

fig1 = plt.figure(figsize=(3.9*cm, 3.9*cm),dpi=600,constrained_layout=True)
ax = fig1.add_subplot(1, 1, 1)
# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(np.cos(sol_fene.t[-(PointsPer+1):]),N1_fene[-(PointsPer+1):],color=[1,0,0],linewidth=linewidth_value)
plt.plot(np.cos(sol_sptt.t[-(PointsPer+1):]),N1_sptt[-(PointsPer+1):],color=[0,0,1],linestyle='dashed',linewidth=linewidth_value)
plt.plot(np.cos(sol_gs.t[-(PointsPer+1):]),N1_gs[-(PointsPer+1):],color=[0,0.85,0],linestyle='dashdot',linewidth=linewidth_value)
# NOTE THE LINE BELOW NEEDS COMMENTING OUT IF THE STRESSES FOR OLDROYD-B MODEL ARE VERY LARGE (I.E. WI/DE IS LARGE).
# plt.plot(np.cos(sol_ob.t[-(PointsPer+1):]),N1_ob[-(PointsPer+1):],color=[1,0.0,1],linestyle='dotted',linewidth=linewidth_value)
plt.xlim([-1.05,1.05])
plt.xticks([-1,1])
plt.ylim([1.05*min(min(N1_fene[-(PointsPer+1):]),min(N1_sptt[-(PointsPer+1):]),min(N1_gs[-(PointsPer+1):]),min(N1_ob[-(PointsPer+1):])),1.05*max(max(N1_fene[-(PointsPer+1):]),max(N1_sptt[-(PointsPer+1):]),max(N1_gs[-(PointsPer+1):]),max(N1_ob[-(PointsPer+1):]))])
plt.yticks([min(min(N1_fene[-(PointsPer+1):]),min(N1_sptt[-(PointsPer+1):]),min(N1_gs[-(PointsPer+1):]),min(N1_ob[-(PointsPer+1):])),max(max(N1_fene[-(PointsPer+1):]),max(N1_sptt[-(PointsPer+1):]),max(N1_gs[-(PointsPer+1):]),max(N1_ob[-(PointsPer+1):]))])
# NOTE THE TWO LINES BELOW NEED UNCOMMENTING IF THE STRESSES FOR OLDROYD-B MODEL ARE VERY LARGE (I.E. WI/DE IS LARGE).
plt.ylim([1.05*min(min(N1_fene[-(PointsPer+1):]),min(N1_sptt[-(PointsPer+1):]),min(N1_gs[-(PointsPer+1):])),1.05*max(max(N1_fene[-(PointsPer+1):]),max(N1_sptt[-(PointsPer+1):]),max(N1_gs[-(PointsPer+1):]))])
plt.yticks([min(min(N1_fene[-(PointsPer+1):]),min(N1_sptt[-(PointsPer+1):]),min(N1_gs[-(PointsPer+1):])),max(max(N1_fene[-(PointsPer+1):]),max(N1_sptt[-(PointsPer+1):]),max(N1_gs[-(PointsPer+1):]))])
# UNCOMMENT THE LINE BELOW IF YOU WANT TO SAVE THE FIGURE.
# plt.savefig("N1_Lissajous_De_" + str(De) + "_Wi_" + str(Wi) + "_theta_" + str(theta) + "_L2_" + str(L2) + ".png")
```
